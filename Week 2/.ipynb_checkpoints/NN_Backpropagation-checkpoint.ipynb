## Perceptron
O Perceptron é uma rede neural com apenas um neurônio. Neste caso, estamos utilizando a rede mais simples possível, com uma única entrada e uma única saída, sem ativação.

Serão utilizados 100 dados para treinar o Perceptron ao longo de 300 épocas.

O objetivo é utilizar o modelo Perceptron para aprender uma simples regressão linear, com o intuito de familiarizar-se com a sintaxe e o funcionamento do TensorFlow.

```python
def f1(x):
    '''
    Funcao a ser aprendida
    '''
    return 5 + 10*x
```

```python
xs = np.linspace(0,10,100)  # gera 100 valores no intervalo [0.10]
ys = f1(xs)                 # computa o valor de f1 nestes 100 valores
#print(len(xs), "xs=", xs)
#print(len(ys), "ys=", ys)
```

```python
#Definindo, compilando e treinando nosso modelo
model = tf.keras.Sequential([
    keras.Input(shape=(1,)),
    keras.layers.Dense(units=1),
])

model.compile(optimizer="sgd", loss="mean_squared_error")
#model.fit(xs,ys,epochs=300)
```

```python
model.summary()
```

```python
print("prediction: "+ str(model.predict([17]))+"      real value: " + str(f1(17)))
```

```python
val = np.linspace(0,10,63)
model.evaluate(x=val, y=f1(val))
```

A função evaluate retorna o "custo" (loss) da avaliação, definido na compilação. Nesse caso, o valor reportado é o erro quadrático médio (MSE).

###### 2.2
Existem várias funções de otimização disponíveis no TensorFlow, e cada uma delas tem sua própria abordagem para atualizar os pesos durante o treinamento. Duas das funções mais comuns são SGD (gradiente descendente estocástico) e RMSprop (média móvel do quadrado dos gradientes).

O SGD é uma abordagem simples e popular que atualiza os pesos proporcionalmente ao gradiente da função de custo. Isso significa que o modelo tenta ajustar os pesos na direção que reduzirá mais rapidamente o erro. No entanto, o SGD pode ser sensível a hiperparâmetros, como a taxa de aprendizagem, e pode não ser capaz de escapar de mínimos locais.

Por outro lado, o RMSprop usa uma abordagem mais sofisticada para ajustar os pesos. Ele mantém uma média móvel dos quadrados dos gradientes anteriores e divide o gradiente atual pelo desvio padrão dessa média móvel. Isso pode ajudar o modelo a evitar mínimos locais e a se mover mais rapidamente em direção ao mínimo global.

Vamos ver como essas duas funções de otimização se comportam em nosso exemplo:

```python
# Teste diferentes funções de otimização
optimizers = ["sgd", "rmsprop"]

for optimizer in optimizers:
    model = tf.keras.Sequential([    
        keras.Input(shape=(1,)),  # Define a camada de entrada com uma dimensão
        layers.Dense(16, activation='relu'),  # Define uma camada densa com 16 unidades e função de ativação relu
        layers.Dense(1)  # Define a camada de saída com 1 unidade
    ])

    model.compile(optimizer=optimizer, loss="mean_squared_error")
    model.fit(x, y, epochs=400)

    model.summary()

    x_val = np.linspace(0, 10, 63)
    y_val = f2(x_val)

    loss = model.evaluate(x=x_val, y=y_val)
    print("Optimizer:", optimizer, "Loss:", loss)
```

Neste exemplo, testamos as funções SGD e RMSprop. Podemos ver que ambas as funções de otimização convergem para um erro baixo depois de um número suficiente de épocas, mas a RMSprop parece ter um desempenho ligeiramente melhor. O desempenho de cada função de otimização também pode ser afetado por outros hiperparâmetros, como a taxa de aprendizagem e o tamanho do lote.

###### 2.3
Não é possível modificar a função de ativação de um algoritmo de otimização, pois ela está relacionada à forma como os neurônios são ativados em cada camada da rede.

Entretanto, podemos modificar o algoritmo de otimização da rede de SDG para Adam e observar o que acontece. O algoritmo de otimização Adam é uma variação do SDG, que usa o conceito de momentos para atualizar os pesos de forma mais eficiente.

Para modificar o algoritmo de otimização da rede de SDG para Adam, basta mudar a linha onde definimos o otimizador no modelo, como no exemplo abaixo:

```python
model = tf.keras.Sequential([
    keras.Input(shape=(1,)),
    layers.Dense(16, activation='relu'),
    layers.Dense(1)
])

optimizer = tf.keras.optimizers.Adam(lr=0.01)
model.compile(optimizer=optimizer, loss='mean_squared_error')
model.fit(x, y, epochs=400)
```

Dessa forma, estamos definindo o otimizador como o Adam, com taxa de aprendizagem (lr) igual a 0.01.

### Questão 3


```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np

def load_data(data_path='data/weatherAUS.csv')-> pd.DataFrame:
    """Funcao que importa dados de um arquivo csv, usando pandas"""
    return pd.read_csv(data_path)

def pre_processing(raw_data:pd.DataFrame)-> pd.DataFrame:
    """Funcao que filtra e limpa os dados meteorologicos para o treinamento"""
    # Seleciona apenas as colunas que serão utilizadas no modelo
    relevant_columns = ['MinTemp', 'MaxTemp', 'Rainfall', 'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm',
                        'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Temp9am', 'Temp3pm', 'RainToday']
    processed_data = raw_data[relevant_columns + ['RainTomorrow']].copy()

    # Remove amostras com valores faltantes
    processed_data.dropna(inplace=True)

    # Converte a coluna 'RainToday' para valores numéricos
    processed_data['RainToday'] = (processed_data['RainToday'] == 'Yes').astype(int)

    # Converte a coluna 'RainTomorrow' para valores numéricos
    processed_data['RainTomorrow'] = (processed_data['RainTomorrow'] == 'Yes').astype(int)

    return processed_data

def split_data(data:pd.DataFrame, val_size= 0.2)-> np.array:
    """Funcao que separa seus dados em teste e treino conforme a proporcao val_size"""
    x = data.drop('RainTomorrow', axis=1)
    y = data['RainTomorrow']
    x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=val_size, random_state=42)

    return x_train, x_val, y_train, y_val
```

```python
df = load_data()
df = pre_processing(df)
visualize_data(df)
x_train, x_val, y_train, y_val = split_data(df,)
```

### Questão 4

```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense

# carrega os dados de precipitação
data = pd.read_csv(r'D:\jupyter\NL-pt\Week 2\data\weatherAUS.csv')

# remove colunas não numéricas
data = data.select_dtypes(include=[np.number])

# separa as colunas de entrada (X) e saída (y)
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values.reshape(-1, 1)

# cria o modelo Sequential
model = Sequential()

# adiciona a camada de entrada
model.add(Dense(units=10, input_dim=X.shape[1], activation='relu'))

# adiciona a camada de saída
model.add(Dense(units=1, activation='sigmoid'))

# compila o modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# treina o modelo
model.fit(X, y, epochs=10, batch_size=32)

# avalia o modelo
scores = model.evaluate(X, y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

Esse código é um exemplo de treinamento de uma rede neural utilizando Keras para classificação binária.

A primeira etapa é carregar os dados de precipitação de um arquivo CSV usando a biblioteca Pandas. Em seguida, as colunas não numéricas são removidas, e as colunas de entrada e saída são separadas em X e y, respectivamente.

A diferença neste exemplo em relação ao código anterior é que a coluna de saída (y) é redimensionada usando o método reshape do NumPy para ter uma coluna adicional, que é necessária para o treinamento da rede neural.

Em seguida, um modelo Sequential é criado e duas camadas são adicionadas: uma camada densa de entrada com 10 unidades e função de ativação ReLU, e uma camada densa de saída com uma única unidade e função de ativação sigmoid.

O modelo é compilado com a função de perda binary_crossentropy, que é usada para classificação binária, e o otimizador adam. A precisão é definida como a métrica de desempenho.

O modelo é treinado por 10 épocas usando um tamanho de lote de 32. Finalmente, o modelo é avaliado usando os dados de treinamento e a precisão é impressa na tela.
